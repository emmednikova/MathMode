Задача 1

В ящике находится 17 качественных и 6 бракованных деталей. Если деталь качественная, вы зарабатываете 10 рублей. Если бракованная, вы отдаете 20 рублей.

Посчитать среднюю прибыль, если достать две детали наугад.

```python
n1 = 17 # качественные детали 
n2 = 6 # бракованные детали 
p1 = 10 # за качественную деталь 10 рублей
p2 = -20 # за бракованную отдаём 20 рублей

# Вероятности достать качественную и бракованную деталь
p_good = n1 / (n1 + n2)
print("Вероятность достать качественную деталь: ", p_good)
p_def = n2 / (n1 + n2)
print("Вероятность достать бракованную деталь: ", p_def)

# Расчет прибыли за каждый вариант достать две детали
profit1 = p1**2 # две детали качественные
profit2 = p1 * p2 # качественная + бракованная
profit3 = p2 * p1 # бракованная + качественная
profit4 = p2**2 # две детали бракованные 

# Средняя прибыль
avg_profit = p_good * p_good * profit1 + p_good * p_def * profit2 + p_def * p_good * profit3 + p_def * p_def * profit4

print("Средняя прибыль, если достать две детали наугад: {} рублей".format(avg_profit))
```

Задача 2

На испытание поставлено 1200 однотипных ламп. За первые 3000 часов работы отказало 70 ламп, а за интервал 3000– 4000 часа отказало еще 50 ламп. Определить частоту и интенсивность отказов электронных ламп в промежутке 3000–4000 часов работы.

```python
N = 1200 # общее количество ламп 
t1 = 3000 # первый интервал времени
t2 = 4000 # второй интервал времени
n1 = 70 # количество отказавших ламп за первый интервал
n2 = 50 # количество отказавших ламп за второй интервал

freq = (n2 - n1) / (N * (t2 - t1)) # частота отказов
inten = (n2 - n1) / (N * (1 - (t1/N))) # интенсивность отказов

print("Частота отказов в промежутке 3000-4000 часов:", freq)
print("Интенсивность отказов в промежутке 3000-4000 часов:", inten)
```

Задача 3

В течение некоторого периода времени производилось наблюдение за работой одного экземпляра радиолокационной станции. За весь период наблюдения было зафиксировано 15 отказов. До начала наблюдения станция проработала 258 часов, а к концу наблюдения наработка станции составила 1233 часа. Требуется определить среднюю наработку на отказ.

```python
start_hours = 258 # до начала наблюдения 
end_hours = 1233 # к концу наблюдения 
failures = 15 # кол-во отказов 

average = (end_hours - start_hours) / failures
print(f"Средняя наработка на отказ: {average} часов")
```

Задача 4

Система состоит из пяти приборов, вероятность исправной работы которых в течение времени t = 100 ч равны: р1(100) = 0,9996; р2(100) = 0,9998; р3(100) = 0,9996; р4(100) = 0,999; р5(100) = 0,9998. Требуется определить частоту отказов системы в момент времени t = 100 ч.

```python
p1 = 0.9996
p2 = 0.9998
p3 = 0.9996
p4 = 0.999
p5 = 0.9998

fail = 1 - (1 - p1) * (1 - p2) * (1 - p3) * (1 - p4) * (1 - p5)
print(f"Частота отказов системы в момент времени t=100 часов: {fail}")
```

Вычислить математическое ожидание выигрыша и округлить его.

Сколько в среднем проигрывает игрок с каждой ставки?
