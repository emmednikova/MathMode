##### Задание 1: При помощи библиотеки NetworkX создайте ненаправленный граф, который состоит из 5 вершин и 4 ребер. Выведите такой граф на экран.

```python
import networkx as nx
import matplotlib.pyplot as plt

# создаём объект графа
G = nx.Graph()
# создаём 5 вершин
G.add_nodes_from([1, 2, 3, 4, 5])
# создаём 4 ребра
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])
# выводим граф на экран
nx.draw(G, with_labels=True, node_size=1000, node_color='skyblue', font_size=15, font_color='black', font_weight='bold')
plt.show()
```

##### Задание 4: При помощи библиотеки NetworkX создайте ненаправленный взвешенный полный граф, у которого 6 вершин (A,B,C,D,E,F). Веса рёбер такого графа пусть имеют значения 1,2,3,4,5 и т.д. по порядку. Выпишите кратчайшие пути между теми вершинами, которые имеют больше одного ребра на своём пути.

```python
# создаём объект графа
Graf = nx.Graph()

# определяем список узлов
nodes = ['A', 'B', 'C', 'D', 'E', 'F']

# определяем список рёбер с весами
weights = {('A', 'B', 2),
              ('C', 'D', 4),
              ('E', 'F', 6),
              ('F', 'A', 1),
              ('E', 'D', 3),
              ('C', 'B', 5)}

# на основе созданных списков вершин и рёбер заполняем граф
# Graf.add_nodes_from(nodes)
Graf.add_weighted_edges_from(weights)

# задаём формат размещения элементов графа для визуализации
#pos = nx.spring_layout(Graf)
pos = nx.circular_layout(Graf)

# выбираем что будет в качестве лейблов (веса)
labels = nx.get_edge_attributes(Graf,'weight')

# выводим граф на экран
nx.draw_networkx(Graf,
        pos,
        with_labels=True,
        node_color='lightblue',
        node_size=500)

# добавляем отрисовку лейблов для рёбер
nx.draw_networkx_edge_labels(Graf, pos, edge_labels=labels)
plt.show()

# определим кратчайший маршрут
shortest = nx.shortest_path(Graf, 'A', 'D', weight='weight')

# определим общее расстояние данного маршрута
len_shortest = nx.shortest_path_length(Graf, 'A', 'D', weight='weight')

print('Кратчайший маршрут', shortest)
print(type(shortest))
print('Маршрут составляет', len_shortest)
print(type(len_shortest))
```
